```python
# Check for missing values
def check_missing_values(dataframe):
    return dataframe.isnull().sum()

print(check_missing_values(df))
df[df.rating_count.isnull()]
```

```python
# Remove rows with missing values in the rating_count column
df.dropna(subset=['rating_count'], inplace=True)
print(check_missing_values(df))
```

```python
# Check data types
def check_data_types(dataframe):
    return dataframe.dtypes

print(check_data_types(df))
```

```python
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%','').astype(float)/100
```

```python
df['rating'] = df['rating'].astype(str).str.replace(',', '').astype(float)
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').astype(float)
```

```python
# Creating the column "rating_weighted"
df['rating_weighted'] = df['rating'] * df['rating_count']
```

```python
df['sub_category'] = df['category'].astype(str).str.split('|').str[-1]
df['main_category'] = df['category'].astype(str).str.split('|').str[0]
```

# 4. Recommendation system

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['user_id_encoded'] = le.fit_transform(df['user_id'])

# Create a new dataframe with the user_id frequency table
freq_table = pd.DataFrame({'User ID': df['user_id_encoded'].value_counts().index, 'Frequency': df['user_id_encoded'].value_counts().values})

# Display the dataframe
print(freq_table)
id_example = freq_table.iloc[0,0]
print(id_example)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def recommend_products(df, user_id_encoded):
    # Use TfidfVectorizer to transform the product descriptions into numerical feature vectors
    tfidf = TfidfVectorizer(stop_words='english')
    df['about_product'] = df['about_product'].fillna('')  # fill NaN values with empty string
    tfidf_matrix = tfidf.fit_transform(df['about_product'])

    # Get the purchase history for the user
    user_history = df[df['user_id_encoded'] == user_id_encoded]

    # Use cosine_similarity to calculate the similarity between each pair of product descriptions
    # only for the products that the user has already purchased
    indices = user_history.index.tolist()

    if indices:
        # Create a new similarity matrix with only the rows and columns for the purchased products
        cosine_sim_user = cosine_similarity(tfidf_matrix[indices], tfidf_matrix)

        # Create a pandas Series with product indices as the index and product names as the values
        products = df.iloc[indices]['product_name']
        indices = pd.Series(products.index, index=products)
        #this new series have product names as keys and indices as values
        # Get the indices and similarity scores of products similar to the ones the user has already purchased
        similarity_scores = list(enumerate(cosine_sim_user[-1]))
        similarity_scores = [(i, score) for (i, score) in similarity_scores if i not in indices]

        # Sort the similarity scores in descending order
        similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

        # Get the indices of the top 5 most similar products
        top_products = [i[0] for i in similarity_scores[1:6]]

        # Get the names of the top 5 most similar products
        recommended_products = df.iloc[top_products]['product_name'].tolist()

        # Get the reasons for the recommendation
        score = [similarity_scores[i][1] for i in range(5)]

        # Create a DataFrame with the results
        results_df = pd.DataFrame({'Id Encoded': [user_id_encoded] * 5,
                                   'recommended product': recommended_products,
                                   'score recommendation': score})

        return results_df

    else:
        print("No purchase history found.")
        return None
```

```python
import pandas as pd
import numpy as np

# Load the existing CSV file with product data
product_df = pd.read_csv("/content/products.csv")

# Extract unique product names, their corresponding prices, ratings, and image links
unique_product_data = product_df[['product_name', 'actual_price', 'rating', 'img_link']].drop_duplicates()

# Generate random user-product pairs and purchase counts
num_users = 1000
min_user_purchases = 5
max_user_purchases = 10

# Create a DataFrame to store user purchase data
user_purchase_data = []

for user_id in range(1, num_users + 1):
    # Generate random purchase counts for the user
    num_purchases = np.random.randint(min_user_purchases, max_user_purchases + 1)
    purchase_counts = np.random.randint(1, 11, size=num_purchases)

    # Randomly select products with replacement
    product_samples = np.random.choice(unique_product_data['product_name'], size=num_purchases, replace=True)
    products = unique_product_data[unique_product_data['product_name'].isin(product_samples)]

    user_purchase_data.extend(zip([user_id] * num_purchases, products['product_name'], products['actual_price'], products['rating'], products['img_link'], purchase_counts))

# Create a DataFrame with the user purchase data
purchase_df = pd.DataFrame(user_purchase_data, columns=['user_id_encoded', 'product_name', 'product_price', 'product_rating', 'img_link', 'purchase_count'])

# Save the user purchase data to a CSV file
purchase_csv_filename = 'user_purchase_data_with_prices_ratings_and_images.csv'
purchase_df.to_csv(purchase_csv_filename, index=False)
print(f'User purchase data with prices, ratings, and images saved to {purchase_csv_filename}')
```

```python
import pandas as pd
import numpy as np

# Load the existing CSV file with user purchase data
purchase_df = pd.read_csv('/content/user_purchase_data_with_prices_ratings_and_images.csv')

# Load the existing CSV file with product data
product_df = pd.read_csv("/content/products.csv")

# Collaborative filtering recommendation function
def item_based_collaborative_filtering_purchase(df, user_id_encoded):
    # Filter user's purchases
    user_purchases = df[df['user_id_encoded'] == user_id_encoded]

    # Find similar users who purchased similar products
    similar_users = df[df['product_name'].isin(user_purchases['product_name'])]['user_id_encoded'].unique()

    # Get products purchased by similar users but not by the input user
    recommended_products = []

    for similar_user in similar_users:
        if similar_user == user_id_encoded:
            continue  # Skip the input user

        similar_user_purchases = df[df['user_id_encoded'] == similar_user]['product_name']
        new_products = set(similar_user_purchases) - set(user_purchases['product_name'])
        recommended_products.extend(new_products)

        if len(recommended_products) >= 5:
            break

    # Randomly select 5 products from the recommended list
    recommended_products = np.random.choice(recommended_products, size=min(5, len(recommended_products)), replace=False)

    # Create a DataFrame with the recommended products, their prices, ratings, and image links
    results_df = pd.DataFrame({
        'Id Encoded': [user_id_encoded] * len(recommended_products),
        'recommended product': recommended_products,
    })

    # Add the prices, ratings, and image links of recommended products to the results DataFrame
    prices = []
    ratings = []
    img_links = []
    for product_name in recommended_products:
        product_info = product_df[product_df['product_name'] == product_name].iloc[0]
        price = product_info['actual_price']
        rating = product_info['rating']
        img_link = product_info['img_link']
        prices.append(price)
        ratings.append(rating)
        img_links.append(img_link)

    results_df['product_price'] = prices
    results_df['product_rating'] = ratings
    results_df['img_link'] = img_links

    return results_df

user_id = 4
recommendations = item_based_collaborative_filtering_purchase(purchase_df, user_id)
print(recommendations)

# Generate collaborative filtering recommendations for each user
# all_recommendations = []

# for user_id in range(1, 1001):
#     recommendations = item_based_collaborative_filtering_purchase(purchase_df, user_id)
#     if not recommendations.empty:
#         all_recommendations.append(recommendations)

# # Combine all recommendations into a single DataFrame
# combined_recommendations = pd.concat(all_recommendations, ignore_index=True)

# # Save the combined recommendations to a new CSV file
# recommendation_csv_filename = 'collaborative_purchase_recommendations_with_ratings_and_images.csv'
# combined_recommendations.to_csv(recommendation_csv_filename, index=False)
# print(f'Recommendations with ratings and images saved to {recommendation_csv_filename}')
```
